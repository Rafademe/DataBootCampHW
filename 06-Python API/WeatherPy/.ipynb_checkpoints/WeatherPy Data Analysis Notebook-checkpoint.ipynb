```python
# Importing necessary libraries and api key
import os
import time
import json
import random
import requests
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from citipy import citipy
from config import owm_api_key
```

```python
# Making an empty list to hold the randomly selected cities
prelim_city_list = []

# Generating 1000 unique cities using randomly generated coordinates
while len(prelim_city_list) < 1000:
    city = citipy.nearest_city(random.randrange(-90,90), random.randrange(-180,180)).city_name
    
    # Ensuring uniqueness in the city list
    if city not in prelim_city_list:
        prelim_city_list.append(city)
```

```python
# Checking number of cities
n_cities = len(prelim_city_list)
print(n_cities)
```

```python
# Creating empty lists to hold weather data for each city and zeroing the loop counter
temp = []
humidity = []
wind = []
cloudiness = []
open_weather_city_number = []
lat = []
lng = []
final_city_list = []
process_number = 0

# Setting up some of the api parameters
url = "api.openweathermap.org/data/2.5/weather?q="
units = "metric"

# Setting up headers for the log (in csv format)
print("Process Number, City Name, City ID")

# Querying the openweathermap api using the cities in the randomly generated list
for city in prelim_city_list:
    if len(final_city_list) < 520:
        query = f"http://{url}{city}&APPID={owm_api_key}&units={units}"
        response = requests.get(query).json()
        process_number = process_number + 1
        time.sleep(1.01) # Delaying by 1.01 secs to stay above the api's 1 query per second limit 
       
        # Attempting to extract each city's temperature, humidity, wind speed, and cloudiness data and filling in the empty lists 
        try:
            temp.append(response['main']['temp'])
            humidity.append(response['main']['humidity'])
            wind.append(response['wind']['speed'])
            cloudiness.append(response['clouds']['all'])
            open_weather_city_number.append(response['id'])
            lat.append(response['coord']['lat'])
            lng.append(response['coord']['lon'])
            final_city_list.append(city)

            # Displaying a log of the last city for which weather data was successfully extracted 
            print(f"{process_number}, {city}, {response['id']}")

        # Displaying a log for the queries that resulted in an error
        except KeyError:
            print(f"{process_number}, {city}, Error: {response['message']}")
```

```python
# Creating a dataframe from the now filled in weather data lists
weather_data_df = pd.DataFrame(data = final_city_list, columns = ["City"])

weather_data_df["Temperature"] = temp
weather_data_df["Humidity"] = humidity
weather_data_df["Wind Speed"] = wind
weather_data_df["Cloudiness"] = cloudiness
weather_data_df["Latitude"] = lat
weather_data_df["Longitude"] = lng
weather_data_df['OpenWeatherMap City Number'] = open_weather_city_number

# Checking the new df
weather_data_df.head(15)
weather_data_df.tail(15)

weather_data_df.describe(include="all").round(3)
```

```python
# Setting up the seaborn style
sns.set()

# Setting up the latitude and longitude plot to vizualize our global coverage
plt.scatter(weather_data_df["Longitude"], weather_data_df["Latitude"])

# Setting titles & labels
plt.suptitle('Latitude & Longitude', verticalalignment="baseline", weight="bold", fontsize=16)
# plt.title(r"$temp \approx temp(lat)$")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Creating a line at the equator
plt.axhline(0, color = 'black', alpha = .25, label = 'Equator')
plt.text(-133,-10,'Equator',)# rotation=90)

# Trimming the axes 
plt.xlim(-186,186)
plt.ylim(-62,82)

# Displaying plot
plt.show()

# Saving the plot
latitude_longitude_plot_path = os.path.join('plots', 'weatherpy_latitude_longitude_plot.png')
plt.savefig(latitude_longitude_plot_path, dpi=300, bbox_inches='tight')
```
