# CBD vs THC

```python
# out of 17233 results, 601 have both thc and cbd information
cannabinoids = prod.drop(columns = ['brand','strain','lab_test'])
pd.options.display.max_rows = 600
clean_cannabinoids = cannabinoids.dropna()
clean_cannabinoids.reset_index(drop = True, inplace = True)
clean_cannabinoids
```

```python
thc = clean_cannabinoids['thc']
cbd = clean_cannabinoids['cbd']
```

```python
print(type(thc))
```

```python
cbd_dominant = []
thc_dominant = []
equal = []

index = len(clean_cannabinoids['thc'])

for i in range(0,index):
    if thc.at[i] == cbd.at[i]:
        equal.append(thc.at[i])
        
    elif thc.at[i] > cbd.at[i]:
        thc_dominant.append(thc.at[i])
        
    else:
        cbd_dominant.append(cbd.at[i])
        
 
```

```python
neutral = pd.DataFrame(equal)
neutral.count()
```

```python
cbd_data = pd.DataFrame(cbd_dominant)
cbd_data.count()
```

```python
thc_data = pd.DataFrame(thc_dominant)
thc_data.count()
```
