## One hot encode the Effect(s) columns in the Dataset- Test

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from numpy import array
from numpy import argmax

# define example
data = final_csv['Effect_1']
values = array(data)
print('Defined Values Output')
print(values)
print('-'*50)

# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print('Integer Encode Output')
print(integer_encoded)
print('-'*50)

# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print('Binary Encode Output')
print(onehot_encoded)
print('-'*50)

# invert first example
inverted = label_encoder.inverse_transform([argmax(onehot_encoded[0, :])])
print('Invert Output')
print(inverted)
```

## The Variable we are predicting,'y', is Rating

```python
# Step 0: Reformat data
data = final_csv['Rating'].values
y = data[:]
y
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
encoded_y
```

```python
# we can observe that each of the original labels has been replaced with an integer
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_y = to_categorical(encoded_y)
one_hot_y
```
